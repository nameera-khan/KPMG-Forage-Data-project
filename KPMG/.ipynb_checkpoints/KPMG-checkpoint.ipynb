# Feature engineering

```python
#calculating recency, frequency and monetary value of customer purchase
#create a snapshot date for recency
from datetime import timedelta
snapshot_date = transactions['transaction_date'].max() + timedelta(days=1)
```

```python
data_process = transactions.groupby(['customer_id']).agg({
        'transaction_date': lambda x: (snapshot_date - x.max()).days,
        'transaction_id': 'count',
        })
```

```python
data_process
```

```python
data_process.rename(columns={'transaction_date':'recency','transaction_id':'frequency'},inplace=True)
```

```python
data_process
```

```python
data_process = pd.merge(sum_of_trans,data_process,on='customer_id')
data_process
```

```python
monetary = data_process['list_price']
```

```python
data_process.drop(columns='list_price',inplace=True)
```

```python
data_process['monetary_value']=monetary
```

```python
data_process
```

```python
# Plot RFM distributions in KDE
plt.figure(figsize=(12,10))
# Plot distribution of R
plt.subplot(3, 1, 1); sns.distplot(data_process['recency'])
# Plot distribution of F
plt.subplot(3, 1, 2); sns.distplot(data_process['frequency'])
# Plot distribution of M
plt.subplot(3, 1, 3); sns.distplot(data_process['monetary_value'])
# Show the plot
plt.show()
```

The plot for recency is positively skewed towards the left while the rest of the 2 variables have a more centralised plot which is good enough.However, the recency plot will need standardisation while modelling the predictions.

By segmenting these features into 4 quantiles which would help in later calculating the RFM score. This score would be used to predict the potential customers and their future purchase decisions. 

```python
#using .qcut() for quantiles in r,f and m groups 
#create labels 
r = range(4,0,-1)
f = range(1,5)
m = range(1,5)

#Assigning labels to quantiles
r_groups = pd.qcut(data_process['recency'], q=4, labels=r)
f_groups = pd.qcut(data_process['frequency'], q=4, labels=f)
m_groups = pd.qcut(data_process['monetary_value'], q=4, labels=m)

#create columns 
data_process = data_process.assign(R= r_groups.values, F = f_groups.values, M = m_groups.values)
```

```python
data_process.head()
```

- 4 denotes the highest or best quantile 

```python
#calculate RFM score
rfm = data_process[['customer_id','recency','frequency','monetary_value','R','F','M']]

rfm['rfm_score']= rfm[['R','F','M']].sum(axis=1)

print(rfm['rfm_score'])
```

```python
rfm
```

These scores can be used to categorise the customer list by RFM level

```python
def rfm_level(df):
    if df['rfm_score'] >= 9:
        return 'Highly potential'
    elif ((df['rfm_score'] >= 6) and (df['rfm_score'] < 7)):
        return 'Loyal'
    elif ((df['rfm_score'] >= 5) and (df['rfm_score'] < 6)):
        return 'Promising'
    elif ((df['rfm_score'] >= 4) and (df['rfm_score'] < 5)):
        return 'Needs Attention'
    else:
        return 'Require Activation'
```

```python
#RFM level categories
rfm['RFM_level']= rfm.apply(rfm_level,axis=1)

rfm.head()
```

```python
#to count the number of customers in each category 
rfm_level_count = rfm.groupby('RFM_level').agg({'recency':'mean',
                                               'frequency':'mean',
                                               'monetary_value':['mean','count']}).round(1)

rfm_level_count
```

The RFM performance shows that about 67% of the customer database regularly purchases from the company. 
- Highly potential - amounting to 37% of the customers who are very beneficial to the business and are considered as the top tier. 
- Loyal- These customers would be placed in the top tier category with just few discrepancies on the frequency of purchase. 
- Promising- These customers show good signs of purchasing in higher number except for the frequency in the transactions. 
- Needs Attention - made very few purchases and hasn't returned. 
- Require activation- They could be potential customers but may have chosen competitor products instead. Given that the monetary value they've added to the businesses is just above the Loyal category. However, distance to travel to the store can also play as a viable reason for this behaviour. This hypothesis can be dealt with further exploratory data analysis

# Applying K Means -Clustering 

To choose optimum number of clusters to predict the customer behaviour

```python
#import sklearn k means clustering module
# range for the clusters 
ks=range(1,12)
inertias=[]

from sklearn.cluster import KMeans

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(rfmt_norm_df)
    inertias.append(model.inertia_) #method to measure inertia 

plt.plot(ks,inertias,'-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

Choosing a point from the 'Elbow cluster chart' would be ideal since it is a good rule of thumb. From the chart above, for our model, k=5 since it is the point where the chart decreases.

```python
model = KMeans(n_clusters = 5)
labels = model.fit_predict(rfmt_norm_df)

#cluster_labels
cluster_labels = model.labels_

#assign the cluster labels to original dataset
rfmt_new = data_process.assign(Cluster = cluster_labels)

rfmt_new.groupby(['Cluster']).agg({
    'recency': 'mean',
    'frequency': 'mean',
    'monetary_value': 'mean',
    'tenure': ['mean', 'count']
}).round(0)
```

Through KMeans prediction the following predictions have been made: 
- Cluster 0 - These customers add premium monetary value to the business with higher purchases in quantity and frequency. It can also be seen that their mean tenure with the company has been for approx. 12 years. 
- Cluster 1 - They are very new customers who have shopped very recently adding monetary value just as much as the top tier premium customers. 
- Cluster 2 - These customers have either gone to competitors or haven't found what they needed in specific in the past one year. 
- Cluster 3 - The findings for this is quite contradictory when monetary value is considered. These customers have shopped very recently, in higher number which does add advantage to the business. 
- Cluster 4- Will need to show more attention to these customers because they may be potential. Perhaps adding options of product customization could add-on to what the customers seek
